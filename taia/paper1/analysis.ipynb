```python
import numpy
import matplotlib.pyplot as plt
```

```python
data = numpy.loadtxt('results_c_100.txt', skiprows=1, usecols=[i for i in range(5) if i not in [0]])
data = data.transpose()
data_default = data[:, ::2]
data_alternative_c = data[:, 1::2]
ncc_default = data_default[2]
ncc_alternative_c = data_alternative_c[2]
cpu_default = data_default[3]
cpu_alternative_c = data_alternative_c[3]

data = numpy.loadtxt('results_100.txt', skiprows=1, usecols=[i for i in range(5) if i not in [0]])
data = data.transpose()
data_default = data[:, ::2]
data_alternative = data[:, 1::2]
ncc_default = data_default[2]
ncc_alternative = data_alternative[2]
cpu_default = data_default[3]
cpu_alternative = data_alternative[3]
```

```python
plt.title("#V = 100")
plt.xlabel('#connected components')
plt.ylabel('cpu seconds')
plt.plot(ncc_default, cpu_default, 'o', label='default')
#plt.plot(ncc_alternative_c, cpu_alternative_c, 'o', label='alternative_c')
plt.plot(ncc_alternative, cpu_alternative, 'o', label='alternative')
plt.legend()
plt.savefig('nV_100.png')
```
